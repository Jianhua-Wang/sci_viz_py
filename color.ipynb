```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import time
# set the font of figure so the text can be edit in Adobe Illustrator
import matplotlib as mpl
mpl.rcParams['font.serif'] = 'Arial'
mpl.rcParams['pdf.fonttype'] = 42
```

# All built-in palettes in Matplotlib

```python
palettes = list(filter(lambda x: False if x.endswith('_r') else True, sorted(mpl.cm.cmap_d.keys())))
palettes.remove('jet')

palettes = pd.DataFrame(data=[sns.color_palette(palette).as_hex() for palette in palettes], index=palettes)

palettes = palettes.fillna('#FFFFFF')

f,ax = plt.subplots(figsize=(12,40))
for y in palettes.index:
    for x in range(20):
        ax.scatter(x=x,y=[y],c=palettes.loc[y,x],s=600,marker='s')
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.get_xaxis().set_visible(False)
```
